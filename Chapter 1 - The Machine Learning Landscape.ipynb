# What is Machine Learning?

**General Definition:** It is the filed of study that gives computers the ability to learn without being explicitly programmed.

**Engineering-oreinted Definition:** A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.

* The examples that the system uses to learn is called the training set.
* Each example in the training set is called a training instance (or sample).
* The part of the system that learns and makes predictions is called a model.
* A performance measure like accuracy is used to then evaluate the effectiveness and effeciency of the model.

# Why use Machine Learning?

Let us consider a spam filter as an example.

When we use a traditional approach to solve the problem we would take the following steps;
* First, we would identify what a spam would look like, words or phrases or other patterns that are common in spam messages.
* Then we would write a detection algorithm for each of the patterns that we identified, the program would then flag the mail as spam if these patterns are detected.
* We would test the program and repeat the above 2 steps until it is good enough to launch.

The issue with this is that we would have to write a long list of complex rules to identify the spam message patterns which is very hard to maintain and update.



In contrast, when we use a machine learning approach, the model automatically learns the words and phrases that are good predictors of spam. The program is also much shorter and easier to maintain.



Now if spammers start to notice that certain words have been identified by the spam filter and marked as spam then they would change these words.

In a traditional way, we would have to write new rules for this and keep updating forever.

But using ML, the model will automatically detect the change and start flagging the changes also as spam without our intervention.



Another area where ML is useful is in problems that are too complex for traditional approaches or have no known algorithm.

It can also be used to dig into large amounts of data to discover patterns that are often hidden to humans and help us learn. This is called as data mining.



# Types of Machine Learning Systems:
* Based on training.
* Based on whether the system can learn incrementally.
* Based on how the model generalize.

## Based on training:

### 1. Supervised Learning:

The training set we feed to the algorithm includes the desired solutions called labels.

A typical supervised ML task is classification. The model predicts the class of a new instance that is given to it for prediction.



Another task of a supervised ML algorithm is to predict a target numeric value, given a set of features. This is called as regression.



### 2. Unsupervised Learning:

The training data is unlabeled.



Here, clustering algorithms are used to detect groups of similar datapoints. New points are grouped to the cluster that is closest to it.



Visualization algorithms are also good examples. A lot of complex and unlabeled data is fed and the output is a 2-D or 3-D representaion of the data. They structure of the data is maintained to identify patterns.



Another related task is dimensionality reduction, in which the goal is to simplify the data without using too much information. This is done by extracting only the important features from the data. This is called feature extraction.

Another important unsupervised task is anomaly detection - detedting and eliminating ouliers and errors.



A related task to the above is novelty detection which aims to detect new instances that look different from all instances in the training set.

Finally, another common unsupervised task is association rule mining. It's goal is to dig into large amounts of data and discover interesting relations between attributes.

### 3. Semi-supervised Learning:

Labeling data is usually a time consuming process. Semi-supervised algorithms can deal with data that is partially labeled. It is basically a combination of supervised and unsupervised learning algorithms. Once the whole dataset is
labeled, any supervised learning algorithm can be used.



### 4. Self-supervised Learning:

Another approach to machine learning involves actually generating a fully
labeled dataset from a fully unlabeled one. Again, once the whole dataset is
labeled, any supervised learning algorithm can be used. This approach is
called self-supervised learning.



### 5. Reinforcement Learning:

The learning system,
called an agent in this context, can observe the environment, select and
perform actions, and get rewards in return (or penalties in the form of
negative rewards). It must then learn by itself what is the best strategy, called a policy, to get the most reward over time. A policy defines what action the agent should choose when it is in a given situation. Applying the policy in a situation is called offline learning.



## Based on whether the system can learn incrementally:

### 1. Batch Learning:

* The system is incapable of learning incrementally.
* It must be trained with all the available data.
* This consumes a lot of time and computing resources.
* The system is trained and launched into production and runs without learning.
* This is called as offline learning.
* Over time, the data undergoes changes which decays the model's perforamnce. This is called as model rot or data drift.
* Solution for this is to regularly retrain the model with up-to-date data.
* Depending on the characteristics of data, the frequency of retraining the model will change.

### 2. Online Learning:

* The system is trianed incrementally.
* The data is fed in instances sequentially or in small groups called as mini-batches.
* Each learning step is fast and cheap, and the system can learn about new data concurrently as it arrives.



* Useful for systems that need to adapt to change rapidly and for those systems that are limited on computing resources.
* It is used to train models on huge datasets that cannot fit in one machine's main memory. This is called as out-of-core learning.
* The alogrithm does the above by loading a part of the data, training with it then loading the next part until all the data is run.



* An important parameter is how fast the model should adapt to changing data. This is called as learning rate.
* High learning rate will allow system to rapidly adapt to new data, and it will tend to forget the old data at the same rate.
* Low learning rate will make the model less sensitive to noise and outliers.
* A challenge for online learning is that if bad data is fed to the system then its performance will drop quickly.
* To solve this we have to constantly monitor the system and we can use anomaly detection algorithms for the data.

## Based on how the model generalize:

### 1. Instance-based learning:

* It is a trivial form of learning and is to simply learn by-heart.
* Creating a spam filter this way would be to simply flag emails similar to that of already flagged emails by the user.
* So a measure of similarity is required between the 2 mails, eg; common words.
* This is called instance-based learning: the system learns the examples by heart, then generalizes to new cases by using a similarity measure to compare them to the learned examples (or a subset of them).



### 2. Model-based learning:

Another way to generalize from a set of examples is to build a model of
these examples and then use that model to make predictions. This is called
model-based learning.



Below is an example of a model-based learning, linear regression is used here:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
# Visualize the data
lifesat.plot(kind='scatter', grid=True,
x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
# Select a linear model
model = LinearRegression()
# Train the model
model.fit(X, y)
# Make a prediction for Cyprus
X_new = [[37_655.2]] # Cyprus' GDP per capita in 2020
print(model.predict(X_new)) # output: [[6.30165767]]
```

If we change the model to a K-nearest neighbors algorithm, we get an instance-based training algorithm:

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X, y)
X_new = [[37_655.2]]
print(model.predict(X_new))
```

# Main challenges of ML:

## Bad Data:

### Insuffecient quantity of training data:
Machine learning algorithms take a lot of data to work properly. Even simple problems require thousands of samples. More complex tasks like image and peech recognition take millions of instances.

### Non-representative training data:
In order for the model to genralize well, it is important that the training data to be represntative of the new cases we want ot generalize to. Non-represntaive datasets will result in inaccurate results. If the sample, is too small then we will have sampling noise. If the sample is too large then we will have sampling bais. Both result in non-represemtative data.

### Poor quality data:
If the data is full of errors, outliers and noise then the system will less likely perform well. So, we need to spend a majority of our time to clean the data. Outliers and noise can be discarded if required. Errors can be fixed manually. Missing values can be filled, attribute dropped entirely or we can train 2 models, one with and the other without that attribute.

### Irrelevent features:
Ths system is cable of performing well only when the training data contains enough relevent features and not too many irrelevent ones. This process is called feature engineering which involves:
* Feature selection - selecting the most useful features to train among all the existing features.
* Feature extraction - combining existing features to prduce more useful ones.
* Creating new features by gathering new data.

## Bad Algorithms:

### Overfitting the training data:
The model performs very well on the training data, but it does not generalize well to new instances. It can occur when the model is too complex relative to the amount of and noisiness of the training data. This can be resolved by, simplifying the model (selecting few attributes or adding constraints to the model), gathering more training data or by reducing the noise in the data.

Constraining the model to make it simpler is called as regularization. If a model take 2 parameters then it is said to have 2 degrees of freedom. The amount of regularization to apply during learning can be controlled by a hyperparameter. The hyperparameter is not affected by the training process, it is constant throughout the whole process.

### Underfitting of training data:
It occurs when the model is too simple to learn the underlying patterns in the data. This can be fixed by selecting a more powerful model, feeding better features or by reducing the constraints of the model.

# Testing and Validating:

The only way to know how well the model will generalize to new cases is to actually try it out on new cases. For this, we split the data into 2: a training set and a testing set. The model is trained using the training set and the it is tested using the testing set. The error rate on the new cases is called the generalization error or out-of-sample error. This value will tell us how well the model will perform to new instances.

## Hyperparameter Tuning and Model Selection:

Suppose we have a hesitation between models, then we would train all and compare how well each generalize using the test set.

Now suppose we have identified model that generalizes the best and we want to generalize it. Then we would need to find the value of the regularization hyperparameter. We could do this by trainig the model using different values of the hyperparameter.

From this we would then select the model that has a hyperparameter value with the lowest generalization error. What if we launch this model and we observe more error than expected?

The problem here is that we adjust the hyperparameter value to find the best fit for that particular dataset only. So, when generalizing it produces poor results.

A solution for this is called holdout validation. We simply houldout a part of the training set to evaluate the candidate models that have already been selected from training using the other part of the training set and select the best one. This new holdout set is called a validation set. After this houldout validation, we train the best model using the full training dataset, this gives the final model. This model can then be tested to get an estimate of the generalization error.



One issue with this is selcting the size of the validation set. It must not be too big or too small. We can solve this using repeated cross-validation, using small validation sets. Each model is evaluated once per validation set. By taking the average of these evaluations we get more accurate measure of its performance. This is however, will increase the training time.

## Data Mismatch:

In some cases we will get a large amount of data for training, but this data won't accurately represent the data that will be used in production. For example we create an app to take a picture of a flower and identify its species. We can download millions of pictures for training from the web, but only a small percentage of them will be similar to the ones that the app would take a picture of.

In this case, both the validation and training set must contain representative data that would be used in production. So, we shuffle the representative pictures and split it equally between the 2 sets. After training however, if the performance is sub-par, we would not know if it is due to the mismatch between the web and mobile pictures or due to some other reason.

One solution for this is to hold out some of the training pictures (from the web) for another set called the train-dev set. After the model is trained on the training set it can be evaluated on the train-dev set. If it performs poorly then its due to other factors (overfitting, over-regularization, unclean data, etc). If it performs well, then it is further evaluated using the dev set. If it performs poorly here, then it is due to data mismatch.

This can be resolved by pre-processing the images from the web such that it looks more like the ones taken by the mobile app. Once the model performs well in the train-dev set and the dev set then it is more likely to perform well in production.



# No Free Lunch Theorem:
For some datasets the best model is a linear model,
while for other datasets it is a neural network. There is no model that is
a priori guaranteed to work better (hence the name of the theorem). The
only way to know for sure which model is best is to evaluate them all.
Since this is not possible, in practice you make some reasonable
assumptions about the data and evaluate only a few reasonable models that adhere to these assumptions.
